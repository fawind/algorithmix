# Aufgabe 1 (Maschinenschaden Scheduling)

Gegeben sind die benötigten Ersatzteile $\{T_1, ..., T_n\}$ mit $T_i = \{r, e\}$, wobei $r$ die Replikationsdauer und $e$ die Einbauzeit ist.

## Algorithmus I

**Ansatz:** Erzeugen der Bauteile aufsteigend nach ihrer Replikationsdauer.

Gegeben folgende Instanz: $T_1 = \{1, 1\}, T_2 = \{2, 1\}, T_3 = \{3, 3\}$. In diesem Fall erzeugt der Algorithmus die Sequenz $T_1, T_2, T_3$, da diese eine aufseigende Replikationsdauer hat. Die gesamte Bearbeitungszeit für diese Sequenz beträgt $9$ Schritte.
Optimal ist jedoch die Sequenz $T_3, T_1, T_2$. In diesem Fall wird das Teil $T_3$ mit der längsten Einbauzeit zuerst produziert, was die gesamte Dauer auf $7$ Schritte verringert. Der Algorithmus ist also nicht optimal.

## Algorithmus II

**Ansatz:** Erzeugen der Bauteile absteigend nach der Summe ihrer Replikations und Einbauzeit.

Gegeben die folgende Instanz: $T_1 = \{10, 1\}, T_2 = \{1, 9\}$. In diesem Fall erzeugt der Algorithmus die Sequenz $T_1, T_2$, da die Gesamtkosten für $T_1$ $11$ und für $T_2$ nur 10 betragen. Insgesamt läuft der Algorithmus $20$ Schritte.
Optimal ist jedoch die Sequent $T_2, T_1$. Hier wird zuerst $T_2$ mit einer Replikationsdauer von $1$ gebaut. Während $T_2$ eingebaut wird, wird dann parallel das aufwendige $T_1$ gebaut. Die Gesamtanzahl der Schritte verringert sich somit auf $11$. Der Algorithmus ist somit nicht optimal.

## Algorithmus III

**Ansatz:** Erzeugen der Bauteile absteigend nach ihrer Einbauzeit.
    
Dieser Algorithmus ist optimal. Betrachtet man nur die Replikationsdauer, wird deutlich, dass ein Tauschen auf Basis der Replikationsdauer keinen Unterschied macht, da die Teile immer nur sequentiell konstruiert werden müssen. Als untere Schranke haben wir somit immer die Summe der Replikationsdauer aller Teile plus die Einbauzeit des zuletzt Replizierten Teiles.  
Da die Einbauzeiten parallel abgearbeitet werden können, ist es in jedem Fall am Besten, die längsten Einbauzeiten zuerst abzuarbeiten.

# Aufgabe 2 (Faule Verdrängungspläne beim Paging)

Unser Verdrängungsplan ist nicht faul, wenn Page $p$ nachgeladen wird, obwohl sie gerade nicht angefragt wird.

**Fall 1:** $p$ wird wieder von einer anderen Page verdrängt, bevor sie benutzt wird.  
In diesem Fall war das Austauschen von $p$ ohne Wirkung. Beim Umwandeln in einen faulen Verdrängungsplan, können wir dieses Nachladen einfach weglassen, wodurch die Anzahl der Nachladungen verringert wird, ohne dass wir zusätzliche Cache-Miss'es bekommen.

**Fall 2:** $p$ wird zu einem späteren Zeitpunkt $t_p$ angefragt.  
In diesem Fall würde es auch reichen, $p$ erst zum Zeitpunkt $t_p$ auszutauschen, da $p$ erst da gebraucht wird. Die Anzahl an Verdrängungen bleibt dabei gleich. Falls am Zeitpunkt $t_p$ bereits eine andere Verdrängung geplant ist, kann diese analog zu den Fällen 1 und 2 entfernt oder auch nach hinten geschoben werden.

Insgesamt lassen sich somit alle Verdrängungspläne in faule Verdrängungspläne umwandeln, wobei die Anzahl der Verdrängungen immer gleich bleibt oder sogar kleiner wird.

# Aufgabe 3 (Beweistraining)

## a) Wir zeigen, dass 2 = 1 gilt.

Dieser Beweis ist ungültig. Aus der ersten Gleichung $a^2 = ab$ wird gefolgert, dass $a = b$. Dadurch ist aber der Schritt, indem durch $(a - b)$ geteilt wird ungültig. Da $a = b$, teilen wir hier durch $0$, was nicht definiert ist.

## b) Wir zeigen, dass alle Autos die selbe Farbe haben.

Dieser Beweis ist ungültig. Der Fehler liegt im Induktionsschritt. Entfernt man ein beliebiges Auto aus $M$, so ist nicht gegeben, dass in der neuen Menge $M^*$ alle Autos gleichfarbig sind. Nach Induktionshypothese, wird $M$ mit $i$ gleichfarbigen Autos und einem anderstfarbigem Auto konstuiert. Wir nun allerdings eins der zuvor gleichfarbigen Autos entfernt, ist $M^*$ nicht gleichfarbig, da das anderstfarbige Auto immer noch in der Menge ist. Auch der Ansatz, dies zweimal zu tun und so die Mengen $M_1$ und $M_2$ zu erhalten, ändert daran nichts. Da der Induktionsschritt falsch ist, ist der Beweis ungültig.

## c) Wir zeigen, dass $\sum^\inf_{n = 1} n = -\frac{1}{12}$ gilt.

Auch dieser Beweis ist ungültig (für reelle Zahlen). Die Reihe ist divergent und hat daher keine Summe. In dem Beweis im Video, limitieren sie die Reihe auf die ersten $n$ partiellen Summen, wobei $n$ gegen unendlich geht (Grandi Summe). Während dies für manche Fälle sinnvoll ist, ist es nicht die eigentliche Summe der Reihe (siehe auch "Eilenberg-Mazur Schwindel").

## d) Wir zeigen, dass 4 = 5 gilt.

Dieser Beweis ist ungültig. Der Fehler liegt im Schritt von Zeile 7 zu 8. $4 - \frac{9}{2}$ ist eine negative Zahl. Da das ziehen von Wurzeln bei negativen Zahlen keine Äquivalenzumformung ist, gilt die Äquivalenz in Zeile 8 nicht mehr.

# Aufgabe 4 (Hoverboards)

## a) Online Algorithmus

**Strategie:** Wir leihen die ersten neun Tage und kaufen am zehnten das Hoverboard.

Sei $t$ die Anzahl der Tage, die die Bahn geöffnet hat. Für $t < 10$ ist die Strategie optimal, da wir im schlimmsten Fall nur die Verleihgebühr für die einzelnen Tage gezahlt haben.

Der worst-case für diesen Algorithmus ist, dass die Bahn erst nach $t \geq 10$ Tagen schließt. In diesem Fall hätten wir 190 anstatt der optimalen 100€ gezahlt. Unser worst-case Faktor ist also $1.9$, da wir im worst-case $1.9$ mal so viel wie das Optimum zahlen.

Würden wir unser Hoverboard erst später kaufen, verschlechtert sich der worst-case Faktor, da wir zusätzliche Leigebühren zahlen müssen, welche man vermeiden könnte. Kaufen wir unsre Hoverboard erst am 11. Tag, ist der worst-case, dass die Bahn am 12. Tag schließt. In diesem Fall hätten wir 200€ gezahlt, wobei die optimale offline Strategie nur 100€ gezahlt hätte. Wir sind also um Faktor $2$ schlechter.

Kennt man den Schließtag nicht, ist es daher am besten, die ersten neun Tage zu mieten, und dann das Board zu kaufen.

## b) Verallgemeinerung

Sei $m$ die Mietkosten und $k$ die Kaufkosten mit $m \leq k$. Wenn die Nutzungstage ungewiss sind, kann man den besten Kauftag $t$ berechnen mit:

$$t = \lceil \frac{k}{m} \rceil$$

Mit anderen Worten: Wir kaufen zu dem Zeitpunkt, an welchem wir zum ersten mal die Gewinnschwelle überschreiten würden.

## c) Beweise, dass es keine bessere deterministische online Strategie gibt.

Der worst-case jeder offline Strategie, ist dass am Tag nach dem Kauf des Hoverboards die Bahn schließt. Insgesamt können wir die gesamten Kosten am Tag $t$ berechnen mit $k(t) = (t - 1) * m + k$, da wir bis zum $t - 1$ten Tag mieten und danach Kaufen.

Im Falle unserer Strategie, tritt der worst-case am Tag $t^* = \lceil \frac{k}{m} \rceil$ ein. In diesem Fall, haben wir $k(t^*) = 2k - m$ ausgegeben, während der optimale Algorithmus nur $k$ ausgegeben hätte. Unsere competitive ratio ist daher $\alpha = \frac{2k - m}{k} = k - \frac{m}{k}$.
Kaufen wir nun zu einem späteren Zeitpunkt $t'$, gilt $t' > t^*$. Dadurch ist gegeben, dass die Gesamtkosten der Strategien, die am Zeitpunkt t' kaufen, immer über den Kosten unserer Strategie bleiben, da die Mietskosten mit der Anzahl der Tage multipliziert werden. Da für alle Fälle mit $t > \frac{k}{m}$ die Kosten des optimalen Algorithmus gleich bleiben, kann die competitive ratio nur größer werden, da der Nenner gleich bleibt, aber der Zähler größer wird.

**Übung:**

Vergleichen mit Kauf am Tag $x$.

Kosten: $(x - 1) * m + k$

**Fall 1:** $x < \lfloor \frac{k}{m} \rfloor$

$\Rightarrow \frac{(x - 1) * m + k}{x * m} = \frac{xm - m + k}{xm} = 1 + \frac{k-m}{xm} > 1 + \frac{k}{k} - \frac{m}{k} = 2 - \frac{m}{k}$

**Fall 2:** $x > \lceil \frac{k}{m} \rceil$

$\Rightarrow \frac{(x - 1) * m + k}{k} = 1 + \frac{xm}{k} - \frac{m}{k} > 2 - \frac{m}{k}$

## d) Analysiere of Randomisierung hilft.

Randomisierung kann eine bessere competitive ratio als unser optimaler deterministischer Algorithmus haben. So könnten wir beispielsweise eine Münze werfen, ob wir an Tag 8 oder wie gewohnt an Tag 10 unser Hoverboard kaufen. Der worst-case tritt auch hier ein, dass nach dem zehnten Tag die Bahn schließt. In diesem Fall hätten wir jedoch nur erwartete Kosten von $0.5 * (70 + 100) + 0.5 * (90 + 100) = 180$. Verglichen mit dem offline Optimum von 100 kommen wir somit auf einen Faktor von $1.8$ anstatt der $1.9$ bei unserem deterministischen online Algorithmus.
