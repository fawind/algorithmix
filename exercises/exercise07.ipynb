# Aufgabe 1

**Algorithmus:**

* Wir starten bei 2.
* Wir fragen ob unsere Zahl größer oder kleiner $n$ ist.
* Ist sie kleiner, vergrößer wir unsere Zahl logarithmisch groß. So lautet die Folge z.b. $2, 2^2, 2^4, 2^{16}, ...$.
* Finden wir so eine Zahl $k$, die größer als $n$ ist, müssen wir nun das Interval $log(k) + 1$ bis $k - 1$ furchsuchen.
* Hierbei starten wir bei $low = log(k) * 2$ und verdoppeln $low$ so lange, bis $low$ auch größer als $k$ ist. Nun wissen wir, dass die Zahl in den Interval $log(k) + 1$ bis $low - 1$ liegen muss.
* Da im schlimmsten Fall unsere Zahl $log(k) + 1$ groß ist, habe wir noch genug Anfgagen, um jede zweite Zahl im neuen Interval anzufragen. Dadurch finden wir garantiert heraus, was unser $n$ ist.

Da wir im ersten Schritt unser Interval logarithmisch vergrößern, machen wir quasi genau das Gegenteil von $\log^*$. Dadurch ist dieser Schritt also in $o(\log^* n)$. Im Zweiten Schritt brauchen wir im schlimmsten Fall auch nur logarithmisch viele Schritte um unser zweites Interval zu bestimmen. Haben wir das zweite Interval, so können wir in $O(1)$ auf unsere gesuchte Zahl schließen. Insgesamt haben wir also eine Laufzeit von $o(\log* n)$

# Aufgabe 2 (Rekurrenzen)

## a) Finden einer geschlossenen Form.

$$ T(n) = 3T(n - 1) + 3n \quad \text{mit } T(1) = 3 $$

### 1. Aufstellen einer Gleichung nach i Iterationen:

**Einsetzen der Gleichung in sich selbst:**

$$\begin{align*}
T(n) =& 3T(n - 1) + 3n \\
=& 9T(n - 2) + 12n - 9 \\
=& 27T(n - 3) + 39n -63 \\
=& 81T(n - 4) + 120n - 306 \\
=& 243T(n - 5) + 363n - 1278
\end{align*}$$

**Zu beobachten:**

1. Der Faktor vor $T(...)$ wird mit $3$ multipliziert ($3^i$).
2. Die Subtraktion von n innerhalb von $T$ vergrößert sich um 1.
3. Der Faktor vor dem $... +xn ...$ erhöht sich in jeder Iteration um $3^i$.
4. Der zu subtrahierende Betrag ($... -x$) erhöht in jeder Iteration um $3^i * (i - 1)$.

Daraus lässt sich folgende Gleichung nach $i$-maligem Einsetzen aufstellen:

$$ T(n) = 3^i T(n - i) + \sum_{j=1}^i {3^j n} - \sum_{j=1}^i {3^j (j - 1)} \tag{1} $$

#### Korrektheitsbeweis:

Beweis per Induktion.

##### Induktionsannahme: $k = 1$

$$ T(n) = 3T(n - 1) + 3n = 3^1 T(n - 1) + \sum_{j=1}^1 {3^j n} - \sum_{j=1}^1 {3^j (j - 1)} $$

Durch unsere Formel erhalten wir das selbe Ergebnis, wie wenn dir die Rekurrenz einmal in sich selber einsetzen. Die Induktionsannahme gilt.

##### Induktionsschritt: $k$ gilt $\Rightarrow$ $k + 1$ gilt:

Angenommen unsere Formel stimmt bis $i = k$. Daraus ergibt sich folgendes:

$$ T(n) = 3^k T(n - k) + \sum_{j=1}^k {3^j n} - \sum_{j=1}^k {3^j (j - 1)} \tag{2} $$

Zu zeigen ist nun, dass wenn wir nun in die obige Formel unsere Rekurrenz einmal einsetzen, wir folgende Formel für $i = k + 1$ erhalten:

$$ T(n) = 3^{k + 1} T(n - (k + 1)) + \sum_{j=1}^{k+1} {3^j n} - \sum_{j=1}^{k+1} {3^j (j - 1)}  \tag{3}$$

Durch einmaliges Einsetzen der Rekurrenz in $(2)$ erhalten wir:

$$\begin{align*}
T(n) =& 3^k T(n - k) + \sum_{j=1}^k {3^j n} - \sum_{j=1}^k {3^j (j - 1)} \\
=& 3^k (3T(n - (k + 1)) + 3(n - k)) + \sum_{j=1}^k {3^j n} - \sum_{j=1}^k {3^j (j - 1)} \\
=& 3^{k+1} T(n - (k + 1)) + 3^{k+1}(n - k) + \sum_{j=1}^k {3^j n} - \sum_{j=1}^k {3^j (j - 1)} \\
=& 3^{k+1} T(n - (k + 1)) + 3^{k+1}n - 3^{k+1}k + \sum_{j=1}^k {3^j n} - \sum_{j=1}^k {3^j (j - 1)} \\
\end{align*}$$

Betrachten wir die beiden Terme am Ende separat:

$$ 3^{k+1}n + \sum_{j=1}^k {3^j n} = \sum_{j=1}^{k+1} {3^j n} \tag{I} $$

$$ -3^{k+1}k - \sum_{j=1}^k {3^j (j - 1)} = - (3^{k+1}k + \sum_{j=1}^k {3^j (j - 1)}) = - \sum_{j=1}^{k+1} {3^j (j - 1)} \tag{II} $$

Setzen wir nun $(I)$ und $(II)$ wieder in unsere Gleichung ein, erhalten wir:

$$ T(n) = 3^{k + 1} T(n - (k + 1)) + \sum_{j=1}^{k+1} {3^j n} - \sum_{j=1}^{k+1} {3^j (j - 1)} $$

Da diese Gleichung $(2)$ entspricht, haben wir gezeigt, dass wenn die Gleichung für $i = k$ gilt, sie auch für $i = k + 1$ gelten muss.

Aus der Induktionsannahme und dem Induktionsschritt fogt, dass die Gleichung $(1)$ die Rekurrenz korrekt nach $i$-maligem Einsetzen beschreibt.

Insgesamt können wir die Rekurrenz $n - 1$ mal in sich einsetzen, bis wir sie direkt lösen können.

### 3. Aufstellen der geschlossenen Form:

$$\begin{align*}
T(n) =& 3^i T(n - i) + \sum_{j=1}^i {3^j n} - \sum_{j=1}^i {3^j (j - 1)} \\

\end{align*}$$

# Aufgabe 3 (Schnelles Sortieren auf möglichst vielen Eingaben)

## a) Es gibt kein vergleichsbasiertes Sortierverfahren, mit Kosten $O(n)$ für die Hälfte aller Eingaben.

Um zu zeigen, dass die Kosten für die Hälfte aller $n!$ vielen Eingaben nicht in $O(n)$ sind, interessiert uns der worst-case der $\frac{n!}{2}$ Eingaben mit den geringesten Kosten. Ist dieser worst-case in $O(n)$, sind folglich alle dieser günstigsten $\frac{n!}{2}$ Eingaben in $O(n)$. Ansonsten ist dies widerlegt, da alle anderen Eingaben höhere Kosten haben müssen.

Nehmen wir an, dass wir maximal $\leq k$ viele Vergleich zum sortieren aller $\frac{n!}{2}$ Eingaben benötigen. Daraus lässt sich schließen, dass wir für alle $\frac{n!}{2}$ Eingaben maximal $\leq 2^k$ viele paarweise verschiedene Vergleiche machen.

Nach dem Taubenschlagprinzip gilt nun, dass wenn $2^k < \frac{n!}{2}$, dann muss der Algorithmus identisch auf zwei verschiedenen Eingaben ausgeführt werden. Daraus folgt, dass der Algorithmus eine der beiden Eingaben falsch bearbeiten muss. Da wir aber davon ausgehen, dass unser Algorithmus korrekt ist, folgt, dass $2^k \geq \frac{n!}{2}$. Nun können wir umformen:

$$
2^k \geq \frac{n!}{2} \\
2 * 2^k \geq n! \\
2^{k+1} \geq (\frac{n}{2})^{(\frac{n}{2})} \\
k + 1 \geq \frac{n}{2} \log_2 \frac{n}{2} \\
k \geq \frac{n}{2} \log_2(\frac{n}{2}) - 1 \notin O(n)
$$

## b) Gibt es ein vergleichsbasiertes Sortierverfahren mit Kosten $O(n)$ für $\frac{1}{2^n}$ aller Eingaben?

Nach dem selben Prinzip, versuchen wir nun für die Ungleichung $2^k \geq \frac{n!}{2^n}$ zu prüfen, ob $k \notin O(n)$ ist.

$$
2^k \geq \frac{n!}{2^n} \\
k \geq \log_2(\frac{n!}{2^n})
$$

Annahme, $k \in O(n)$:

$$
\log_2(\frac{n!}{2^n}) \in O(n) \\
\Leftrightarrow \frac{n!}{2^n} \in O(2^n) \\
\Leftrightarrow n! \in O(2^{2n})
$$

Dies ist jedoch falsch, da $n!$ viel schneller wächst, als $2^{2n}$. Die Annahme $k \in O(n)$ muss daher falsch sein.

## c) Wie groß ist der Anteil aller Eingaben, auf denen in Kosten $O(n)$ sortiert werden kann?

Gesucht ist eine Funktion $f(n)$, sodass $\log(\frac{n!}{f(n)}) \in O(n)$ gilt.

$$
\log(\frac{n!}{f(n)}) \in O(n) \\
\frac{n!}{f(n)} \leq 2^n \\
\frac{n!}{2^n} \leq f(n)
$$

Wir können also maximal einen Anteil von $\frac{1}{f(n)}$ mit $f(n) \leq \frac{n!}{2^n}$ in Kosten $O(n)$ sortieren.

# Aufgabe 4 (Insertionsort, Mergesort, Quicksort)

## a) Laufzeit von Insertionsort in $O(n + |I(\sigma)|)$.

Die äußere for-Schleife immer $n - 1$ Iterationen läuft, ist sie in $O(n)$.

Finden wir innerhalb der for-Schleife einen Fehlstand, so tasuchen wir in der inneren while-Schleife das kleinere Element des Fehlstands immer weiter an den Anfang, bis kein Fehlstand mehr vorliegt. Die innere Laufzeit ergibt sich also aus der Anzahl der Vertauschungen, bis keine Fehlstände mehr vorliegen. Da bei jedem Tausch in der while-Schleife ein Paar $(i, j)$ mit $i < j$ und $\sigma(i) > \sigma(j)$ vertauscht werden, wird dadurch ein Fehlstand aufgelößt, da anschließen das kleinere der beiden Elemente einen kleineren Index als das größere Element hat. Zudem kann durch diesen Tausch auch kein neuer Fehlstand entstehen.  
Da bei jeder Iteration der inneren while-Schleife ein Fehlstand beseitigt, benötigt die innere while-Schleife also maximal $O(|I(\sigma)|)$ viele Schritte um alle Fehlstände aufzulösen.

Damit liegt die insgesamte Laufzeit in $O(n + |I(\sigma)|)$.

## b) Rekurrenz für die Kosten von Mergesort.

Angenommen $n$ ist eine Zweierpotenz, dann wird bei jedem rekursiven Aufruf von `Mergesort` $n$ die Länge der Eingabe halbiert. Zusätzlich werden in jeder Rekursion, zwei weitere rekursive Aufrufe gemacht. Daraus folgt, dass $a = 2$ und $b = 2$ sein muss (Faktor der rekursiven Aufrufe und Verkleinerungsfaktor der Eingabe). Weiterhin sind die Kosten für Merge in $O(n)$. Somit können wir folgende Rekurrenz aufstellen:

$$ T(n) = 2T(\frac{n}{2}) + n \quad \text{mit } T(1) = 1 $$

Nach Fall $2$ des Master-Theorems gilt:

$$ f(n) \in \Theta(n^{\log_2 2}) = \Theta(n) $$

## c) Rekurrenz für die Kosten von Quicksort.

Wenn als Pivot immer das kleinste Element im aktuellen Teilarray gewählt wird, so wird im worst-case bei einem falsch rum sortieren Array, der Pivot immer als größter Index im aktuellen Teil-Array gewählt. In diesem Fall, müssen alle anderen Felder mit dem Pivot vertauscht werden ($O(n)$ Aufwand). Nach `Partitioniere` befindet sich der Pivot dann am Index $0$. Dies hat zur Folge, dass nur der zweite rekursive Aufruf von `Quicksort` ausgeführt wird (der erste Aufruf bricht direkt ohne weitere Kosten ab, da $r \leq l$). Der nächste Aufruf von Qucksort hat zudem die Eingabegröße $n - 1$, da sich der Pivot nun am Anfang des Teil-Arrays befindet. Aus dieser pessimistischen Abschätzung, lässt sich die Folgende Rekurrenz bilden:

$$ T(n) = T(n - 1) + n, \quad \text{mit } T(1) = 1 $$

Nach $i$-maligem Einsätzen, erhält man:

$$ T(n) = T(n - i) + in $$

Es gilt $i = n - 1$ als maximale Aufrufe bis zum Rekursionsabbruch, da sich unser Teil-Array in jeder Iteration um eins verkleinert. Setz man dies ein, so erhält man die geschlossene Form:

$$ T(n) = T(1) + (n - 1) n = n^2 - n + 1 \in O(n^2) $$
